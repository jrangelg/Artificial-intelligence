# What is an statistic model?

A stochastic model is a mathematical model that incorporates randomness or uncertainty into its structure. It is used to represent systems or processes that exhibit probabilistic behavior. In a stochastic model, various aspects of the system are described by random variables or probability distributions.

## Examples

1. $y_i = f(x_i) + ϵ_i$

2. $y_i = f(x_i)  ϵ_i$

3. $y_i \sim ( \mu(x_i), \sigma^2(x_i)  )$

Unlike deterministic models, which produce the same output for a given set of inputs, stochastic models introduce randomness to account for factors that are difficult to predict or quantify precisely. These models recognize that many real-world phenomena are influenced by random variables or chance events.

Stochastic models are widely used in fields such as statistics, mathematics, physics, economics, finance, engineering, and computer science. They can be employed to study a wide range of phenomena, including stock market fluctuations, population dynamics, weather patterns, traffic flow, queueing systems, and more.

# Geometric Brownian Motion

We consider calssical ODE

$$
dx(t) = \mu * x(t)dt
$$

We can solve this for separable variables

$$
x(t)= x_0 \ e^{\mu t}
$$

Now we will see exponential growth as a random variable, so the equation that generalizes this modeling is the following:
\begin{equation*}
    dX_t = \mu X_t dt + \sigma X_t dBt
\end{equation*}
The second addend is used to balance the effect of the quadratic variation, giving us the following process as a result:
\begin{equation*}
    X_t = x_0 \ \text{exp}\left[ \left( \mu -\frac{1}{2}\sigma^2\right)t + \sigma B_t  \right].
\end{equation*}
Now let's show that it is a random generalization of classical exponential growth:
\begin{equation*}
    \begin{split}
        E(X_t) &= E \left( x_0 \ \text{exp}\left[ \left( \mu -\frac{1}{2}\sigma^2\right)t + \sigma B_t  \right] \right) \\
        &= x_0 \ \text{exp}\left[ \left( \mu -\frac{1}{2}\sigma^2\right)t \right]E \left( \text{exp}[\sigma B_t] \right) \\
        &= x_0 \ \text{exp}\left[ \left( \mu -\frac{1}{2}\sigma^2\right)t \right] \text{exp}\left[\frac{1}{2} t \sigma^2 \right] \\
        &= x_0 \ e^{\mu t}
    \end{split}
\end{equation*}

# When to use Geometric Brownian Motion
Geometric Brownian Motion (GBM) is a mathematical model commonly used to describe the behavior of financial assets, particularly stock prices. It assumes that the logarithmic returns of the asset over time follow a normal distribution with constant mean and standard deviation.

GBM is widely used in various areas of finance and economics, and here are a few situations where it can be applied:

1. Stock price modeling: GBM is often used to model stock prices and simulate their future behavior. It assumes that the underlying asset's price changes continuously and randomly, making it suitable for short-term price predictions and option pricing models like the Black-Scholes model.

2. Risk analysis: GBM is used in risk management and portfolio optimization to model the uncertainty and volatility associated with financial assets. By simulating various scenarios using GBM, analysts can estimate the potential risks and make informed decisions to manage their portfolios.

3. Derivative pricing: Many derivative instruments, such as options, are priced using models based on GBM assumptions. The Black-Scholes model, which assumes constant volatility and log-normal distribution of stock prices, is a famous example that uses GBM to calculate option prices.


```python
import numpy as np
import matplotlib.pyplot as plt

def geometric_brownian_motion(T, N, mu, sigma, S0):
    dt = T/N
    t = np.linspace(0, T, N+1)
    W = np.random.standard_normal(size=N+1)
    W = np.cumsum(W)*np.sqrt(dt)  # Movimiento Browniano estándar
    X = (mu-0.5*sigma**2)*t + sigma*W
    S = S0*np.exp(X)  # Proceso de Browniano Geométrico
    return t, S

# Parámetros de la simulación
T = 1   # Tiempo final
N = 500  # Número de pasos
mu = 2  # Tasa de crecimiento esperada
sigma = 0.2  # Volatilidad
S0 = 5    # Precio inicial

# Simulación de trayectorias
num_trayectorias = 10
trayectorias = []
for _ in range(num_trayectorias):
    t, S = geometric_brownian_motion(T, N, mu, sigma, S0)
    trayectorias.append(S)

# Gráfica de las trayectorias
plt.figure(figsize=(10, 6))
for i in range(num_trayectorias):
    plt.plot(t, trayectorias[i])
plt.plot(t,S0*np.exp(mu*t))
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Simulation of Geometric Brownian Motion')
plt.grid(True)
plt.show()
```

GBM has two main parameters: the expected rate of return and volatility. These parameters are needed to calibrate the model and make predictions or perform risk analysis. Here are some common ways to estimate the parameters of GBM:

Historical estimation: Historical data of the financial asset can be used to calculate the average rate of return and volatility. The rate of return can be estimated as the average of past returns, while volatility can be calculated using the standard deviation of past returns. However, it's important to note that parameters estimated using historical data may not necessarily reflect future movements of the asset.

Parametric estimation methods: Statistical methods can be used to estimate the parameters of GBM. For example, the Maximum Likelihood Estimation (MLE) method can be employed to find parameter values that maximize the likelihood of observing the historical data. There are also other parametric estimation methods such as the method of moments that can be used to estimate the parameters of GBM.

Estimation methods based on options: Since GBM is widely used in options pricing, option price data can be used to infer the parameters of the model. Option pricing models like the Black-Scholes model can be used to back out the parameters of GBM from observed option prices.

It's important to note that estimating the parameters of GBM is challenging, and the results can be influenced by the quality of the available data and underlying assumptions. Additionally, GBM has limitations and cannot capture all the characteristics of real asset price behavior.

# Ornstein Uhlenbeck Process

The Ornstein-Uhlenbeck process is a stochastic process commonly used in various fields, including finance, physics, and biology. Here are a few situations where the Ornstein-Uhlenbeck process is often applied:

1. Financial modeling: The Ornstein-Uhlenbeck process is used to model the mean-reverting behavior observed in many financial time series. It is particularly useful for modeling the dynamics of interest rates, exchange rates, and asset prices that tend to revert towards a long-term mean. The process captures the notion of mean reversion, where the current value of a variable is influenced by both its long-term mean and its deviation from that mean.

2. Option pricing: The Ornstein-Uhlenbeck process is employed in option pricing models to account for stochastic volatility. For instance, the Heston model, which is widely used to price options, combines the Ornstein-Uhlenbeck process for volatility with the underlying asset price process to capture the volatility clustering observed in financial markets.

3. Physics and engineering: The Ornstein-Uhlenbeck process is used in physics and engineering to describe the random motion of particles subject to frictional forces. It has applications in areas such as Brownian motion, diffusion processes, and random noise modeling.

4. Biology and neuroscience: The Ornstein-Uhlenbeck process is utilized to model the stochastic dynamics of various biological systems, such as the movement of organisms, neuronal activity, and gene expression. It can capture the tendency of biological processes to fluctuate around an equilibrium state.

5. Signal processing and noise modeling: The Ornstein-Uhlenbeck process is employed in signal processing to model and filter noisy signals. It provides a way to characterize the temporal correlation structure of the noise and can be used to remove unwanted noise components from signals.

It's important to note that the Ornstein-Uhlenbeck process assumes several assumptions, such as stationarity, Gaussianity, and linear mean reversion. These assumptions may not hold in all situations, and it is crucial to assess their validity and consider alternative models when necessary.

$$
dX_t = \theta X_t dt + \sigma dB_t
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_ornstein_uhlenbeck(n_paths, n_steps, dt, theta, mu, sigma, x0):
    paths = np.zeros((n_paths, n_steps+1))
    paths[:, 0] = x0

    for i in range(n_paths):
        for j in range(n_steps):
            paths[i, j+1] = paths[i, j] + theta * (mu - paths[i, j]) * dt + sigma * np.sqrt(dt) * np.random.normal()

    return paths

# Parameters
n_paths = 50
n_steps = 100
dt = 0.01
theta = 0.5
mu = 1.0
sigma = 0.2
x0 = 0.0

# Simulate trajectories
paths = simulate_ornstein_uhlenbeck(n_paths, n_steps, dt, theta, mu, sigma, x0)

# Plot trajectories
time = np.arange(0, n_steps+1) * dt
for i in range(n_paths):
    plt.plot(time, paths[i])

plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Ornstein-Uhlenbeck Process - 50 Trajectories')
plt.show()
```
