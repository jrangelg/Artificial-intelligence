```r
# Cargar librerías
library(ggplot2)
library(GGally)

# Función para añadir contornos en la parte inferior (nubes)
lower_fun <- function(data, mapping, ...) {
  ggplot(data = data, mapping = mapping) +
    geom_point(alpha = 0.8, size = 1.5, ...) +
    geom_density2d(aes(color = Species), size = 0.6) +
    theme(panel.grid = element_blank())
}

# Crear la matriz con bordes y contornos
ggpairs(
  iris,
  columns = 1:4,
  aes(color = Species),
  upper = list(continuous = wrap("cor", size = 3)),
  lower = list(continuous = lower_fun),
  diag = list(continuous = wrap("densityDiag"))
) +
  theme_minimal() +
  theme(
    panel.border = element_rect(color = "blue", fill = NA, size = 0.7),  # Borde en cada celda
    strip.background = element_rect(fill = "gray90", color = "black")
  )
```

# Regresión lineal

$$
Y_i = \beta_0 + \beta_1 x_i + ɛ_i
$$

$$
ɛ_i \sim Normal
$$

Homegeneidad

Independencia

Si se cumplen los tres supuestos se puede realizar inferencia sobre la significancia de la pendiente

```r
attach(iris)
mod<- lm(Petal.Width~Petal.Length)
```

# Supuestos sobre los errores

## Normalidad

$$
H_0: ɛ \sim NORMAL
$$
$$
H_a: ɛ \not \sim NORMAL
$$

```r
install.packages('nortest')
library(nortest)
lillie.test(mod$residuals)
```

Bajo un nivel de significancia del 0.05 se rechaza la normalidad.

Bajo un nivel de significancia del 0.01 no se rechaza la normalidad.

## Homogeneidad - Homocedasticidad

$$
H_0: Homocedasticidad
$$
$$
H_a: Heterocedasticidad
$$

```r
install.packages("lmtest")
library(lmtest)
bptest(mod)
```

```r
plot(mod)
```

```r
# Independencia

$$
H_0: Independencia
$$
$$
H_a: Dependencia
$$

```

```r
dwtest(mod)
```

```r
attach(iris)
mod <- lm(Petal.Width ~ Petal.Length + I(Petal.Length^2), data = iris)
```

```r
lillie.test(mod$residuals)
```

```r
bptest(mod)
```

```r
dwtest(mod)
```
