# ACTIVO CON RENTA VARIABLE MODELADA BAJO BROWNIANO GEOMÃ‰TRICO

```python
# prompt: Simule un browniano geomÃ©trico con plot 100 trayectorias

import numpy as np
import matplotlib.pyplot as plt

# ParÃ¡metros del modelo
S0 = 100  # Precio inicial
mu = 0.5  # Tasa de crecimiento promedio
sigma = 0.2  # Volatilidad
T = 1  # Tiempo de vencimiento
dt = 0.01  # Paso de tiempo
n_paths = 100  # NÃºmero de trayectorias

# Generar trayectorias
t = np.arange(0, T + dt, dt)
paths = np.zeros((n_paths, len(t)))
paths[:, 0] = S0

for i in range(1, len(t)):
  dW = np.random.normal(0, np.sqrt(dt), n_paths)
  paths[:, i] = paths[:, i - 1] * np.exp((mu - 0.5 * sigma ** 2) * dt + sigma * dW)

# Graficar las trayectorias
plt.figure(figsize=(10, 6))
for i in range(n_paths):
  plt.plot(t, paths[i, :])

plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.title("SimulaciÃ³n de Browniano GeomÃ©trico")
plt.grid(True)
plt.show()
```

```python
# Graficar las trayectorias
plt.figure(figsize=(10, 6))
plt.plot(t, paths[1, :])

plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.title("SimulaciÃ³n de Browniano GeomÃ©trico")
plt.grid(True)
plt.show()
```

```python
u=np.zeros((n_paths, len(t)-1))
for i in range(n_paths):
  for j in range(len(t)-1):
    u[i,j]=np.log(paths[i,j+1]/paths[i,j])
```

```python
# Graficar las trayectorias
plt.figure(figsize=(10, 6))
for i in range(n_paths):
  plt.plot(t[:-1], u[i, :])  # Use t[:-1] to match the length of u[i, :]

plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.title("SimulaciÃ³n de Browniano GeomÃ©trico")
plt.grid(True)
plt.show()
```

```python
est_sig=[]
for i in range(n_paths):
  est_sig.append(np.std(u[i,:])/np.sqrt(dt))
```
