## Introducción

Cuando se desea realizar un estudio, lo que se busca compartir con la comunidad científica son:

1) Observaciones relevantes obtenidas en la práctica o el entorno de estudio.

2) Datos específicos recolectados en grupos con características comunes.

3) Diferencias significativas que se identifican al comparar grupos.

En cualquiera de estos escenarios, el cálculo del tamaño de muestra respalda el propósito fundamental de toda investigación, el cual generalmente se orienta a uno de dos.  

**Objetivos**:

a) Estimar un parámetro de interés

b) contrastar una hipótesis.



**¿Por qué?**

- Limitaciones de tiempo, costos y acceso a todos los individuos.
- Riesgos éticos o logísticos al intervenir o recopilar información de manera innecesaria.

Por esto, el muestreo se convierte en una herramienta esencial para obtener conclusiones válidas a partir de una parte representativa de la población.

## ¿Por qué usamos muestras y no poblaciones completas?

### Razones clave

1. **Rapidez:** Se pueden recolectar y analizar los datos más fácilmente y en menos tiempo.

2. **Costo:** Estudiar menos unidades (personas, empresas, objetos, etc.) es mucho más económico.

3. **Viabilidad:** Muchas veces no es posible estudiar a toda la población por cuestiones logísticas o prácticas.

4. **Precisión:** Cuando el muestreo y la recolección de datos se hacen correctamente, los resultados obtenidos a partir de muestras pueden ser igual de válidos y confiables que los de un censo.

> **Una buena muestra = resultados válidos sin estudiar a todos.**

Esto permite obtener conclusiones útiles y tomar decisiones informadas en cualquier área de investigación, optimizando recursos y tiempo.


## ¿Por qué es necesario calcular el tamaño de muestra?

Imagina que deseas realizar una investigación para responder preguntas clave sobre una población:  
- ¿Cuál es la prevalencia de un problema de salud?  
- ¿Qué factores se asocian al rendimiento académico en colegios?  
- ¿Funciona una nueva estrategia de ventas en una empresa?

Si la muestra es **muy pequeña**, corres el riesgo de no detectar diferencias o relaciones importantes (baja potencia estadística).  
Si la muestra es **excesivamente grande**, puedes gastar recursos innecesarios y exponer a más personas o unidades de estudio de las que realmente se requieren.

Por eso, **calcular el tamaño de muestra adecuado** es el paso que conecta tus objetivos de investigación con un diseño eficiente y riguroso.

### Ejemplo práctico
> Un hospital quiere saber si un nuevo tratamiento reduce el tiempo de recuperación de pacientes.  
> ¿Cuántos pacientes debe incluir en el estudio para detectar una diferencia relevante?  
> Si usa muy pocos, podría no ver la diferencia. Si usa demasiados, podría estar usando recursos de más.

---

Recuerda!!  
**El tamaño de muestra adecuado asegura que tus resultados sean confiables, válidos y eficientes, maximizando el uso de recursos y la validez científica.**

---




Un buen diseño requiere **planear el tamaño apropiado de la muestra** antes de iniciar el trabajo de campo o la recolección de datos.

Este proceso se conoce como la **determinación de la potencia del estudio** (o "power" en inglés), que se expresa como $ 1 - \beta $.  

La **potencia estadística** es la capacidad de un estudio para detectar una diferencia (o un efecto) de un tamaño determinado, en caso de que realmente exista.

- Un estudio con **baja potencia** puede no detectar diferencias importantes (alto riesgo de error tipo II).
- Un estudio con **potencia adecuada** permite identificar diferencias reales de manera confiable y evita desperdiciar recursos.

> **Planificar el tamaño de muestra y la potencia del estudio es fundamental para asegurar la validez y la utilidad de los resultados científicos.**


## Consideraciones Previas: Nivel de Investigación

Antes de calcular el tamaño de muestra, es fundamental definir el **nivel de investigación** que se realizará. Esto orienta el diseño, el tipo de datos a recolectar y los métodos estadísticos a emplear.

- **Nivel I (Exploratorio):**  
  Investigación preliminar que busca obtener una visión general de un fenómeno. Ejemplo: encuestas iniciales, estudios de prevalencia.

- **Nivel II (Descriptivo):**  
  Se enfoca en describir características o comportamientos de una población o fenómeno específico. Ejemplo: estudios demográficos, análisis de frecuencias.

- **Nivel III (Correlacional):**  
  Analiza la relación o asociación entre dos o más variables. Ejemplo: estudios que buscan identificar factores asociados a una condición.

- **Nivel IV (Explicativo):**  
  Busca determinar causas y efectos, usualmente a través de experimentos o ensayos controlados. Ejemplo: ensayos clínicos, estudios experimentales en cualquier campo.

---

> **Definir el nivel de investigación ayuda a seleccionar el tipo de muestreo, el tamaño de muestra y el análisis estadístico más apropiado.**


## Consideraciones Previas: Tipo de Variable

Antes de elegir la fórmula para el cálculo del tamaño de muestra, es necesario identificar el **tipo de variable principal** en el estudio:

- **Variables cualitativas:**  
  Pueden ser **nominales** (categorías sin orden, como sexo, tipo de empresa, estado civil) u **ordinales** (categorías con orden, como niveles educativos o de satisfacción).  
  Estas variables suelen analizarse usando **proporciones** o **frecuencias**.

- **Variables cuantitativas:**  
  Pueden ser **discretas** (número de hijos, cantidad de productos vendidos) o **continuas** (peso, altura, ingresos, tiempo).  
  Estas variables suelen analizarse a través de **medias**, **varianzas** y otras medidas numéricas.

> **Recomendación:** Siempre que sea posible, mantén las variables en formato cuantitativo, ya que permiten extraer y analizar más información estadística.

---

**¿Por qué importa?**  
La **fórmula para calcular el tamaño de muestra** depende del tipo de variable principal que se analizará (proporciones para cualitativas, medias para cuantitativas).




## Datos Necesarios para el Cálculo del Tamaño de Muestra

Para calcular el tamaño de muestra adecuado se requieren tres elementos principales:

- **Variabilidad**  
  - Para proporciones: valores esperados de **p** (proporción de éxito) y **q** (proporción de fracaso), donde $q = 1 - p$.
  - Para medias: **s** (desviación estándar) estimada de la variable principal.
  - Estos valores se obtienen de literatura previa, estudios piloto o estimaciones conservadoras (por ejemplo, usar p = 0.5 si no se tiene información).

- **Nivel de confianza deseado (Uso de Z)**  
  - Es el grado de certeza que se quiere tener sobre los resultados.
  - Comúnmente se usan 90% (Z = 1.64), 95% (Z = 1.96) o
 99% (Z = 2.58).

- **Precisión absoluta (\(d\)) o relativa  ($\varepsilon$)**  
  - **d:** Es el margen de error máximo tolerable, referido a la amplitud del intervalo de confianza para el parámetro que se desea estimar.
  
  - **($\varepsilon$):** Puede usarse como margen de error relativo, en proporción al valor esperado.

---

> **Nota:**  
> Estos valores se deben definir antes del cálculo, idealmente usando antecedentes, literatura científica, estudios piloto o, en su defecto, valores conservadores que aseguren la validez del estudio.

---

**Ejemplo:**  
- Si quieres estimar la proporción de usuarios satisfechos con un servicio, pero no tienes datos previos, puedes usar **p = 0.5** para maximizar la variabilidad y así calcular el tamaño de muestra más conservador.



```r
install.packages(c("pwr", "ggplot2", "plotly"))
```

```r
# Cargar librerías
library(pwr)
library(ggplot2)
library(plotly)

# Parámetros configurables
d <- 0.5                 # Tamaño del efecto (Cohen's d)
alpha <- 0.05            # Nivel de significancia
powers <- seq(0.6, 0.99, by = 0.01)

# Cálculo del tamaño de muestra para cada valor de poder
tamaños <- sapply(powers, function(p) pwr.t.test(d = d, power = p, sig.level = alpha, type = "two.sample")$n)
df <- data.frame(Poder = powers, Tamaño_muestra = tamaños)

# Gráfico con ggplot2
p <- ggplot(df, aes(x = Poder, y = Tamaño_muestra)) +
  geom_line(color = "blue", size = 1) +
  geom_point(color = "black") +
  labs(
    title = "Relación entre el poder estadístico y el tamaño de muestra",
    subtitle = paste("Prueba t para dos muestras independientes (d =", d, ", alfa =", alpha, ")"),
    x = "Poder estadístico",
    y = "Tamaño de muestra requerido (n)"
  ) +
  theme_minimal()

p
```

## Fórmulas para calcular la muestra en estudios descriptivos

### A) Estudios con variable cuantitativa

#### 1. ¿Cuándo usar cada fórmula?

- **Población infinita (N > 1m):**
  Se asume que la población es muy grande o desconocida, por lo que el tamaño de muestra no requiere corrección por población finita.

 $n = \frac{Z^2 \cdot S^2}{d^2}$

- **Población finita (N ≤ 1m):**
  Se conoce el tamaño total de la población y este no es muy grande, por lo que es necesario corregir el tamaño de muestra para evitar sobreestimarlo.

  
  $n = \frac{N \cdot Z^2 \cdot S^2}{d^2(N-1) + Z^2 \cdot S^2}$

---


#### 2. Claves para el uso correcto



- **Siempre estima o busca un valor razonable para S² (varianza poblacional estimada) usando datos previos, estudios piloto o literatura.**
- **Elige d (precisión) acorde a la necesidad del estudio.**  
  Un error más pequeño (d) exige muestras más grandes.

- **El valor de Z depende del nivel de confianza:**  

  - 90% → Z = 1.64  
  - 95% → Z = 1.96  
  - 99% → Z = 2.58

- **Si tienes dudas sobre el tamaño de la población, usa la fórmula de población infinita (es más conservadora).**



```r
# Parámetros
Z <- 1.64       # Nivel de confianza (95%)
S <- 12           # Desviación estándar estimada (por ejemplo)
d <- 2            # Precisión absoluta deseada
N <- 8000         # Tamaño total de la población

# Fórmula para población infinita
n_inf <- (Z^2 * S^2) / (d^2)

# Fórmula para población finita
n_finite <- (N * Z^2 * S^2) / ((d^2 * (N - 1)) + (Z^2 * S^2))

cat("Tamaño de muestra (población infinita):", ceiling(n_inf), "\n")
cat("Tamaño de muestra (población finita):", ceiling(n_finite), "\n")
```

## Tamaño de muestra usando error relativo $\varepsilon$

En vez de definir la precisión absoluta (d), puedes definir el **error relativo** como un porcentaje de la media esperada ($\bar{X}$), es decir:


$\varepsilon$ = $\frac{d}{\bar{X}}$


**Fórmulas:**

- **Población infinita:**

    $n = \frac{Z^2 \cdot S^2}{(\varepsilon \cdot \bar{X})^2}$
  

- **Población finita:**
  
  $n = \frac{N \cdot Z^2 \cdot S^2}{(\varepsilon \cdot \bar{X})^2 (N - 1) + Z^2 \cdot S^2}$

---

### **Código base en R**





```r
# Parámetros
Z <- 1.64          # Nivel de confianza (95%)
S <- 12               # Desviación estándar estimada
media <- 50           # Media esperada (por ejemplo, de estudios previos)
epsilon <- 0.03      # Error relativo permitido (5%)
N <- 8000            # Tamaño de la población

# Calcula el margen de error absoluto
d <- epsilon * media

# Fórmula para población infinita
n_inf <- (Z^2 * S^2) / (d^2)

# Fórmula para población finita
n_finite <- (N * Z^2 * S^2) / ((d^2 * (N - 1)) + (Z^2 * S^2))

cat("Tamaño de muestra (población infinita):", ceiling(n_inf), "\n")
cat("Tamaño de muestra (población finita):", ceiling(n_finite), "\n")
```

## Estudios con variable cualitativa

En estudios con variables cualitativas (proporciones), se calcula el tamaño de muestra según la proporción esperada del fenómeno a estudiar.

### Fórmulas

- **Población infinita (N > 10,000):**
  
  $n = \frac{Z^2 \cdot p \cdot q}{d^2}$

- **Población finita (N ≤ 10,000):**
  
  $n = \frac{N \cdot Z^2 \cdot p \cdot q}{d^2(N-1) + Z^2 \cdot p \cdot q}$

Donde:
- p = proporción esperada del fenómeno
- q = 1 - p
- d = precisión absoluta deseada



```r

# Parámetros
Z <- 1.96           # Nivel de confianza (95%)
p <- 0.5            # Proporción esperada (usar 0.5 si se desconoce)
q <- 1 - p
d <- 0.05           # Precisión absoluta deseada (por ejemplo, 5%)
N <- 1200           # Tamaño total de la población

# Población infinita
n_inf <- (Z^2 * p * q) / (d^2)

# Población finita
n_finite <- (N * Z^2 * p * q) / ((d^2 * (N - 1)) + (Z^2 * p * q))

cat("Tamaño de muestra (población infinita):", ceiling(n_inf), "\n")
cat("Tamaño de muestra (población finita):", ceiling(n_finite), "\n")
```

## Fórmulas para calcular la muestra en estudios explicativos

En estudios **explicativos** (por ejemplo, ensayos clínicos, estudios de intervención, estudios comparativos o de asociación), el cálculo del tamaño de muestra requiere considerar:

1. **Tipo de hipótesis a contrastar:**  
   - ¿La prueba será **unilateral** (se busca un cambio solo en una dirección) o **bilateral** (se busca cualquier diferencia, mayor o menor)?

2. **Riesgo aceptado de cometer error tipo α $(\alpha)$:**  
   - Es el nivel de significancia estadística, generalmente 0.05 (5%), es decir, la probabilidad de detectar una diferencia cuando no existe (falso positivo).

3. **Riesgo aceptado de cometer error tipo β $(\beta)$ y potencia estadística $(1-\beta)$:**  
   - La **potencia** es la probabilidad de detectar una diferencia cuando realmente existe. Usualmente se busca una potencia de 80% o 90% $\beta = 0.20$ o **0.10**.

4. **Magnitud mínima de la diferencia o efecto a detectar:**  
   - ¿Cuál es la diferencia mínima entre grupos (o asociación) que se considera **importante detectar**? (por ejemplo, diferencia de medias, proporciones, odds ratio, etc.).

5. **Variabilidad de respuesta en el grupo de referencia:**  
   - Para variables continuas: desviación estándar estimada.
   - Para proporciones: valor esperado en el grupo control o referencia.

---

> Todos estos elementos son esenciales para definir el tamaño de muestra adecuado y lograr resultados estadísticamente y clínicamente relevantes.

---

**Nota:**  
La fórmula específica dependerá del tipo de comparación (dos proporciones, dos medias, odds ratio, riesgos relativos, etc.), pero siempre necesitarás estos cinco elementos para hacer el cálculo.




## Cálculo de muestra para prueba de hipótesis en una proporción de población

El objetivo es determinar cuántos sujetos necesitas para detectar si la proporción de un evento o característica en la población **difiere** de un valor de referencia, con una probabilidad (potencia) y error definidos.

### **Información necesaria:**
- **P₀:** Proporción bajo la hipótesis nula (valor de referencia)
- **P₁:** Proporción esperada en la población (bajo la hipótesis alternativa)
- **α:** Nivel de significancia (probabilidad de error tipo I, usualmente 0.05)
- **β:** Probabilidad de error tipo II (1 - potencia, normalmente β = 0.20 para potencia del 80%)
- **Tipo de hipótesis:**  
  - **Unilateral:** Esperas un aumento o disminución específica (P₁ > P₀ o P₁ < P₀)
  - **Bilateral:** Cualquier cambio (P₁ ≠ P₀)

---

### **Fórmula para prueba unilateral:**


$n = \left(
\frac{Z_\alpha \sqrt{P_1(1-P_1)} + Z_\beta \sqrt{P_0(1-P_0)}}
{P_1 - P_0}
\right)^2$

- $Z_\alpha$: valor crítico de Z para el nivel de significancia (α)
- $Z_\beta$: valor crítico de Z para la potencia deseada (1-β)

---



```r

# Parámetros
P0 <- 0.30          # Proporción bajo hipótesis nula (por ejemplo, valor histórico)
P1 <- 0.45          # Proporción esperada a detectar (efecto mínimo relevante)
alpha <- 0.05       # Nivel de significancia
beta <- 0.10        # Error tipo II (para potencia del 90%)

Z_alpha <- qnorm(1 - alpha)         # Unilateral: cola superior
Z_beta <- qnorm(1 - beta)           # Potencia: cola superior

# Fórmula
n <- ((Z_alpha * sqrt(P1 * (1 - P1)) + Z_beta * sqrt(P0 * (1 - P0))) / (P1 - P0))^2

cat("Tamaño de muestra necesario (prueba unilateral):", ceiling(n), "\n")
```

## Cálculo de muestra para comparar dos medias

Cuando tu estudio busca detectar una **diferencia entre las medias de dos grupos independientes**, se usa la siguiente aproximación, bajo estos **supuestos**:

- Las desviaciones estándar de ambos grupos son iguales.
- Los tamaños de muestra serán iguales en ambos grupos.

Se utiliza la **prueba t de Student** para dos muestras independientes.

---

### **Información necesaria:**

- **Diferencia mínima de interés entre medias:**  
  - $\mu_1 - \mu_0$  
- **Desviación estándar común $(\sigma)$** de ambas poblaciones (estimada de estudios previos o piloto).
- **Nivel de significancia $(\alpha)$** (típicamente 0.05).
- **Potencia estadística $(1-\beta)$** (típicamente 0.80 o 0.90).
- **Hipótesis alternativa**:  
  - Generalmente bilateral $(P_a \ne P_0)$, pero puede ser unilateral si tienes dirección esperada.

---

### **Fórmula para tamaño de muestra en cada grupo:**

La fórmula para calcular el tamaño de muestra por grupo al comparar dos medias independientes es:

$$
n = 2 \left( \frac{(Z_{\alpha} + Z_{\beta}) \cdot \sigma}{\varepsilon} \right)^2
$$

Donde:  
- $n$ = tamaño de muestra por grupo  
- $Z_{\alpha}$ = valor Z para el nivel de significancia  
- $Z_{\beta}$ = valor Z para la potencia deseada  
- $\sigma$ = desviación estándar común  
- $\varepsilon$ = diferencia mínima de interés entre las medias





```r
alpha <- 0.05
beta <- 0.10
sigma <- 10
delta <- 5

Z_alpha <- qnorm(1 - alpha/2) # para bilateral
Z_beta <- qnorm(1 - beta)

n <- 2 * ((Z_alpha + Z_beta) * sigma / delta)^2
ceiling(n) # Tamaño por grupo
```

# fórmula para calcular el tamaño de muestra en estudios de casos y controles (usando odds ratio) es:

$$
n = \frac{Z^2_{1-\alpha/2} \left[ \frac{1}{P_1^*(1 - P_1^*)} + \frac{1}{P_2^*(1 - P_2^*)} \right]}
      {[\log_e(1 - \varepsilon)]^2}
$$

Donde:

- $n$ = tamaño de muestra por grupo (casos y controles)
- $Z_{1-\alpha/2}$ = valor crítico de Z para el nivel de confianza bilateral
- $P_1^*$ = probabilidad de exposición en enfermos (casos)
- $P_2^*$ = probabilidad de exposición en no enfermos (controles)
- $\varepsilon$ = precisión relativa (por ejemplo, 0.20 para una precisión del 20%)
- $OR$ = odds ratio esperado (necesario para derivar $P_1^*$ y $P_2^*$ si solo se conoce $OR$ y una de las proporciones)

**Recuerda:**  
Para utilizar esta fórmula, necesitas estimar $P_1^*$ y $P_2^*$ a partir de literatura, estudios piloto, o haciendo supuestos sobre la prevalencia de exposición y el OR esperado.



### **Cálculo del tamaño de muestra en estudios de casos y controles (Odds Ratio)**

Supongamos:
- Probabilidad de exposición en casos: $P_1^* = 0.30$
- Probabilidad de exposición en controles: $P_2^* = 0.15$
- Nivel de confianza: 95%  ($\alpha = 0.05$, $Z_{1-\alpha/2} = 1.96$)
- Precisión relativa deseada: $\varepsilon = 0.20$  (20%)



```r

# Parámetros
P1 <- 0.30         # Probabilidad de exposición en casos (enfermos)
P2 <- 0.15         # Probabilidad de exposición en controles (no enfermos)
alpha <- 0.05
epsilon <- 0.20

Z <- qnorm(1 - alpha/2)    # Z para 95% de confianza (bilateral)

# Cálculo según la fórmula
numerador <- Z^2 * (1 / (P1 * (1 - P1)) + 1 / (P2 * (1 - P2)))
denominador <- (log(1 - epsilon))^2

n <- numerador / denominador

cat("Tamaño de muestra necesario por grupo (casos y controles):", ceiling(n), "\n")
```

### Cálculo de muestra en estudios de cohortes (riesgo relativo)

En estudios de cohortes, el tamaño de muestra puede calcularse para detectar una diferencia específica en el **riesgo relativo (RR)** entre expuestos y no expuestos, con una precisión relativa $\varepsilon$.

**Información necesaria:**
- $P_1$: Probabilidad prevista de enfermedad (u evento) en expuestos
- $P_2$: Probabilidad prevista de enfermedad en no expuestos
- $RR$: Riesgo relativo esperado ($P_1 / P_2$)
- $\alpha$: Nivel de significancia (típico: 0.05)
- $\varepsilon$: Precisión relativa (por ejemplo, 0.20 = 20%)

**Fórmula:**

$$
n = \frac{Z^2_{1-\alpha/2} \left[ \frac{1 - P_1}{P_1} + \frac{1 - P_2}{P_2} \right]}
{[\log_e(1 - \varepsilon)]^2}
$$

---

#### **Ejemplo en R**

Supón:
- Probabilidad de evento en expuestos: $P_1 = 0.18$
- Probabilidad de evento en no expuestos: $P_2 = 0.12$
- Nivel de confianza: 95% ($\alpha = 0.05$)
- Precisión relativa: $\varepsilon = 0.20$




```r

# Parámetros
P1 <- 0.18
P2 <- 0.12
alpha <- 0.05
epsilon <- 0.20

Z <- qnorm(1 - alpha/2)  # Z para 95% bilateral

numerador <- Z^2 * ((1 - P1) / P1 + (1 - P2) / P2)
denominador <- (log(1 - epsilon))^2

n <- numerador / denominador

cat("Tamaño de muestra necesario por grupo (cohortes):", ceiling(n), "\n")
```

### Cálculo de muestra con coeficiente de correlación de Pearson

Cuando el objetivo del estudio es **detectar una asociación lineal** entre dos variables cuantitativas (correlación de Pearson), se utiliza la siguiente fórmula:

$$
N = \left( \frac{Z_\alpha + Z_\beta}{0.5 \cdot \ln\left(\frac{1+r}{1-r}\right)} \right)^2 + 3
$$

Donde:
- $N$: Tamaño de muestra necesario
- $r$: Coeficiente de correlación esperado en la población
- $Z_\alpha$: Valor Z para el nivel de significancia (usualmente bilateral)
- $Z_\beta$: Valor Z para la potencia deseada
- $\ln$: Logaritmo natural (neperiano)

---

#### **Ejemplo **

Supongamos:
- Correlación esperada: $r = 0.30$
- Nivel de significancia: $\alpha = 0.05$
- Potencia: $1-\beta = 0.80$




```r

# Parámetros
r <- 0.1
alpha <- 0.05
beta <- 0.10

Z_alpha <- qnorm(1 - alpha/2) # Bilateral
Z_beta <- qnorm(1 - beta)

den <- 0.5 * log((1 + r) / (1 - r))
N <- ((Z_alpha + Z_beta) / den)^2 + 3

cat("Tamaño de muestra necesario para detectar r =", r, "es:", ceiling(N), "\n")
```

#### Caso 8: Tamaño de muestra para comparación de medidas pareadas

En estudios antes/después (por ejemplo, medición basal y post-intervención en los mismos sujetos), la fórmula se basa en la diferencia esperada y la desviación estándar de las diferencias.

Fórmula:
$$
n = \left( \frac{(Z_{\alpha/2} + Z_\beta) \cdot \sigma_d}{\delta} \right)^2
$$

Donde:
- $\sigma_d$: desviación estándar de las diferencias individuales
- $\delta$: diferencia mínima esperada entre las dos mediciones




```r

alpha <- 0.05
beta <- 0.20
sigma_d <- 8   # desviación estándar de las diferencias
delta <- 5     # diferencia mínima de interés

Z_alpha <- qnorm(1 - alpha/2)
Z_beta <- qnorm(1 - beta)

n <- ((Z_alpha + Z_beta) * sigma_d / delta)^2
ceiling(n)
```

### **9. ANOVA (comparación de más de dos medias)**

 Tamaño de muestra para ANOVA (comparar más de dos grupos)

Fórmula (simplificada, grupos iguales):
$$
n = \frac{2 \cdot (Z_{\alpha} + Z_\beta)^2 \cdot \sigma^2}{k \cdot \delta^2}
$$

Donde:
- $k$: número de grupos
- $\sigma^2$: varianza común estimada
- $\delta$: diferencia mínima relevante entre medias de los grupos




```r

alpha <- 0.05
beta <- 0.20
sigma <- 6
delta <- 4
k <- 3

Z_alpha <- qnorm(1 - alpha/2)
Z_beta <- qnorm(1 - beta)

n <- (2 * (Z_alpha + Z_beta)^2 * sigma^2) / (k * delta^2)
ceiling(n)
```

# Clasificación de los métodos de muestreo

Los métodos de muestreo se dividen en dos grandes grupos:

1. **Muestreo probabilístico:**  **Estudios cuantitativos**
   Cada elemento de la población tiene una probabilidad conocida y diferente de cero de ser seleccionado.

    Diseños de muestreo: Entorno R   (Universo Finita)
    - MAS: Muestreo Aleatorio Simple
    - Estratificado
    - Conglomerados

    Inferencia: Representación, Aleatorización, Cantidad adecuada de elementos, Generalizacion para el universo de estudio


2. **Muestreo no probabilístico:**  **Estudios cualitativos**

   No todos los elementos tienen probabilidad conocida o igual de ser elegidos. La selección depende de factores como la accesibilidad, el juicio del investigador o la facilidad de contacto.


# Muestreo No Probabilístico

La selección no está basada en probabilidad, sino en factores prácticos o juicio del investigador. Es más rápido y económico, pero **no permite generalizar** con confianza a toda la población.

Principales tipos:
- **Conveniencia**

- **Juicio o Intencional**
Experto quien selecciona quien esta en la muestra.

- **Por Cuotas**: Estratificado, cuando se separa la poblacion por grupos, no hay criterio seleccion.


- **Bola de Nieve**: Poblaciones raras, estudios específicos para seleccionar.


# Muestreo por Conveniencia

- **Definición:** Se eligen los elementos que son más fáciles de encontrar o acceder.
- **Ventaja:** Rápido y económico.
- **Limitación:** Alto riesgo de sesgo, poca representatividad.

**Ejemplo:** Encuestar a personas en un centro comercial.

Muestreo Aleatorio Simple sin Reemplazo (MAS)
El Muestreo Aleatorio Simple sin Reemplazo (MAS) es una de las técnicas de muestreo probabilístico más utilizadas. Consiste en seleccionar una muestra de tamaño n de una población de tamaño N, de tal manera que todas las muestras posibles tienen la misma probabilidad de ser seleccionadas.

Características principales:
Cada individuo tiene la misma probabilidad de ser seleccionado.

Una vez seleccionado un individuo, no puede volver a ser seleccionado (sin reemplazo).

Es el diseño más simple y suele ser la base para comparar otras técnicas de muestreo.

Ventajas:
Fácil de entender y aplicar.

Resultados fácilmente extrapolables a la población bajo supuestos de homogeneidad.

Limitaciones:
Requiere un listado completo de la población.

No es eficiente si la población es muy grande o está dispersa.
