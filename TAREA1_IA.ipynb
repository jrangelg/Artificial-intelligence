```python
import pandas as pd

class CustomIterable:
    def __init__(self, file_path):
        self.data = pd.read_csv(file_path)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.data):
            raise StopIteration("error")

        next_data = self.data.iloc[self.index:self.index + 2]
        self.index += 2

        return next_data



```

```python
import pandas as pd
import matplotlib.pyplot as plt

class CustomDataAnalyzer:
    def __init__(self, file_path):
        self.data = pd.read_csv(file_path)

    def calculate_basic_statistics(self):
        return self.data.describe()

    def available_variables(self):
        return self.data.columns

    def analyze_variable(self, variable_name):
        if variable_name not in self.data.columns:
            return "Variable not found"

        variable_stats = self.data[variable_name].describe()

        plt.figure(figsize=(10, 6))
        plt.hist(self.data[variable_name], bins=20, edgecolor='black')
        plt.title(f"Distribution of {variable_name}")
        plt.xlabel(variable_name)
        plt.ylabel("Frequency")
        plt.show()

        return variable_stats

# Ruta al archivo CSV en tu Google Drive
file_path = '/content/drive/MyDrive/IAyNL/Iris.csv'

# Instanciar la clase con el conjunto de datos
data_analyzer = CustomDataAnalyzer(file_path)

# Obtener estadísticas descriptivas básicas
basic_stats = data_analyzer.calculate_basic_statistics()
print("Estadísticas descriptivas básicas:")
print(basic_stats)

# Obtener variables disponibles
variables = data_analyzer.available_variables()
print("\nVariables disponibles:")
print(variables)

# Analizar una variable específica
variable_name = 'sepal.length'
variable_stats = data_analyzer.analyze_variable(variable_name)
print(f"\nEstadísticas de la variable {variable_name}:")
print(variable_stats)
```
